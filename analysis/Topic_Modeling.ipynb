```python
# %load /Users/piorkja1/Documents/PycharmProjects/Platform_Access/RedditTopicModeler.py

__author__ = 'piorkja1'

import json
import re
import gensim

# setup for tokenization and stopwords
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
from gensim import corpora, models, similarities
#from collections import defaultdict

stop_words = set(stopwords.words('english'))
stop_words.update(["@usedgov", "@foxnews", "@cnn", "@msnbc", "@nytimes", "@npr"])


def is_not_num(x):
    try:
        int(x)
        return False
    except:
        return True

f = open("../data/text.csv", "r")

all_tokens = []
for tweet in f.readlines():
    parsed_tweet = []
    for i in wordpunct_tokenize(tweet):
        i = re.sub(r'[^\w\s]','',i)
        i = i.lower()
        if i not in stop_words:
            parsed_tweet.append(i)
    all_tokens.append(" ".join(parsed_tweet))

# all_tokens = ' '.join(words)
# texts = words
print(0)
# remove words that appear only once
# tokens_once = set(words for words in set(all_tokens) if all_tokens.count(words) == 1)

# texts = [[words for words in texts ]
#          for words in all_tokens]
print(all_tokens)


# Setup gensim dictionary
dictionary = corpora.Dictionary()
print(1)

corpus = [dictionary.doc2bow(text) for text in all_tokens]
corpora.MmCorpus.serialize('twitter.mm', corpus)
print(2)
lda = gensim.models.LdaModel(corpus, id2word=dictionary, alpha='auto', num_topics=10)
for i in lda.show_topics():
    print (i)

```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
pyLDAvis.enable_notebook()

# feed the LDA model into the pyLDAvis instance
topic_vis = gensimvis.prepare(lda, corpus, dictionary)

# topic_vis = pyLDAvis.gensim.prepare(lda, corpus, dictionary)

pyLDAvis.display(topic_vis)

```
